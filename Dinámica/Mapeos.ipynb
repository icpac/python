```python
"""Puse dos transformaciones complejas
1 / z y 1 / zconjugada que cumplen con: |z| |1/z| = 1
|z||1/ zconjugada| = 1

Y con ellas dibujé unos círculos y un triángulo."""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#función 1 / z
def funcion(x, y):
    return x / (x**2 + y**2), -y / (x**2 + y**2) 


#Circulo
t = np.linspace(0, 2*np.pi, 200)

plt.figure(figsize=(10, 7))
plt.plot(np.sin(t), np.cos(t), alpha=.2)
plt.annotate("Unitario", (0, 1), alpha=.5)


#Circulo interior
xs = 1/3*np.sin(t)+1/2
ys = 1/3*np.cos(t)
plt.plot(xs, ys)
plt.annotate("sin transformar", (-.5, 0.4), alpha=.5)

#Circulo transformado
nxs, nys = funcion(xs, ys)
plt.plot(nxs, nys)
plt.annotate("transformado", (3.8, 2), alpha=.5)

plt.title("Mapeo 1 / z")
plt.show()
```

```python
#funcion 1/ z conjugada
def funcion(x, y):
    return x / (x**2 + y**2), + y / (x**2 + y**2) 

#dominio
t = np.linspace(0, 2*np.pi, 60)


plt.figure(figsize=(6, 5))
plt.plot(np.sin(t), np.cos(t), alpha=0.2)
plt.annotate("Unitario", (0, 1), alpha=.5)

#Triángulo dentro del circulo unitario
#primer segmento
cntsPnts = 20
segx = np.linspace(-7/8, -1/2, cntsPnts, endpoint=True)
segy = np.linspace(1/10, -1/2, cntsPnts, endpoint=True)
plt.plot(segx, segy, c="r")
plt.annotate("original", (-.5, 0), alpha=.5)

#Segmentro transformado
nxs, nys = funcion(segx, segy)
plt.plot(nxs, nys, c="b")


#Segundo segmento
segx = np.linspace(-1/2, -1/4, cntsPnts, endpoint=True)
segy = np.linspace(-1/2, -1/4, cntsPnts, endpoint=True)
plt.plot(segx, segy, c="r")

nxs, nys = funcion(segx, segy)
plt.plot(nxs, nys, c="b")


#Tercer segmento
segx = np.linspace(-1/4, -7/8, cntsPnts, endpoint=True)
segy = np.linspace(-1/4, 1/10, cntsPnts, endpoint=True)
plt.plot(segx, segy, c="r")

nxs, nys = funcion(segx, segy)
plt.plot(nxs, nys, c="b")
plt.annotate("transformado", (-2.1, .1), alpha=.5)

plt.title("Mapeo 1 / z conjugada")
```

```python
# funcion 1/z
def funcion(x, y):
    return x / (x**2 + y**2), -y / (x**2 + y**2) 

#dominio
t = np.linspace(0, 2*np.pi, 60)

xs = 1/3*np.sin(t)-3/2
ys = 1/3*np.cos(t)

nxs, nys = funcion(xs, ys)

#circulo unitario
plt.figure(figsize=(7, 5))
plt.plot(np.sin(t), np.cos(t), alpha=0.2)
plt.annotate("Unitario", (0, 1), alpha=.5)

#círculo original fuera del unitario
plt.plot(xs, ys)
plt.annotate("original", (-1.67, .40), alpha=.5)

#transformado, adentro del unitario
plt.plot(nxs, nys)
plt.annotate("tranformado", (-.9, .23), alpha=.5)

plt.title("Mapeo 1 / z")
plt.show()
```

```python
import pandas as pd
import unicodedata
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
#, encoding = "ISO-8859-1"
df = pd.read_csv("Homicidios2.csv", encoding = "utf-8")
print(df)
```

```python
print(df["Ano"])
print(df[" Total"])
```

```python
data = [go.Bar(x=df["Ano"] , y=df[" Total"])]
layout = go.Layout(title="Homicidios por año",
                   xaxis= dict(title="Años"),
                   yaxis= dict(title="Homicidios totales"),
                   barmode = "stack" )
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

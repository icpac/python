```python
import matplotlib.pyplot as plt
import numpy as np
from fractions import Fraction
```

```python
"""
Función Doubling
2x
2x-1
2x mod 1
"""
def Doubling(x):
    return 2*x % 1
```

```python
def DoublingFrac(x):
    y = 2*x % 1
    return Fraction(y).limit_denominator()
```

```python
def Dibuja():
    """ Se dibuja la función Doubling """
    dom = np.arange(0, 0.5, 0.0156) 
    plt.plot(dom, Doubling(dom), color="b")
    dom = np.arange(0.5, 1, 0.0156)
    plt.plot(dom, Doubling(dom), color="b")
    
```

```python
def DibujaIden():
    """ Se dibuja la función identidad """
    dom = np.arange(0, 1, .01)
    plt.plot(dom, dom, "y--")
```

```python
Dibuja()
DibujaIden()
```

```python
""" Se calcula la órbita de val 
y se imprimen """
def Orbita(val, niter):
    y = val
    for i in range(niter):
        print(f"{y} ; ", end="")
        y = Doubling(y)
```

```python
""" Se puede observar que se va 
perdiendo precisión """
Orbita(1/5, 20)
```

```python
Orbita(1/9, 20)
```

```python
def OrbitaSin(val, niter):
    """Se calcula la órbita sin perder 
    precisión """
    y = Fraction(val).limit_denominator()
    for i in range(niter):
        print(f"{y} ; ", end="")
        y = Fraction(Doubling(y)).limit_denominator()
```

```python
OrbitaSin(1/5, 30)
```

```python
OrbitaSin(1/9, 40)
```

```python
def DibujaTel(x0, nItera, funcion, xi, xf):
    """Diagrama de Telaraña """
    itera = 1
    y = funcion(x0)
    plt.plot([x0, x0], [x0, y], color="pink")
    while y > xi and y < xf and itera < nItera:
        plt.plot([x0, y], [y, y], color="red")
        x0 = y
        y = funcion(x0)
        plt.plot([x0, x0], [x0, y], color="pink")
        itera +=1
```

#Diagrama de Telaraña sin precisión

```python
Dibuja()
DibujaIden()
DibujaTel(1/5, 256, Doubling, 0, 1)
```

#Diagrama de Telaraña sin perder precisión

```python
Dibuja()
DibujaIden()
DibujaTel(1/5, 256, DoublingFrac, 0, 1)
```

```python
Dibuja()
DibujaIden()
DibujaTel(1/9, 256, Doubling, 0, 1)
```

```python
Dibuja()
DibujaIden()
DibujaTel(3/22, 512, DoublingFrac, 0, 1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xdi = -2
xdf = 7
paso = 0.01

x = np.arange(xdi, xdf, paso)

funcion = lambda t: t**2-4*t+1
y = funcion(x)
```

```python
plt.plot(x, x, color="blue", label="y = x")
plt.plot(x, y, color="green", label="y = 4x(1-x)")
```

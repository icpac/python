```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 2*np.pi, 70)

def xt(t):
    return (1+2*np.cos(t))*np.cos(t)

def yt(t):
    return (1+2*np.cos(t))*np.sin(t)


#plt.xticks([])
#plt.yticks([])
plt.grid(True)  


plt.plot(xt(t), yt(t))
plt.title("limacon")
```

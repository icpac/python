```python
"""Se usa 'solve', para resolver una ecuación: encontrar x tal que 
    haga cero a la ecuación. Es decir se busca su intersección con el eje X"""
```

```python
from sympy import symbols, Eq, solve
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = symbols('x')

solc = solve(3*x + 2)
print("Solución: ", solve(3*x + 2, x))
```

```python
print(solc, np.array(solc)*0)
```

```python
x = np.arange(-5, 4, 0.1)

plt.axhline(y=0, color='k', alpha=.1)
plt.axvline(x=0, color='k', alpha=.2)

plt.title("3*x + 2")
plt.plot(x, 3*x+2)
plt.annotate("Solución: (-2/3, 0) ", (-3, 1.5), alpha=.5)

if len(solc) > 0:
    plt.plot(solc, np.array(solc)*0, 'o', color='black')
```

```python
x = symbols('x')

solc = solve(3*x**2 - 2*x - 5)
print("Soluciónes: ", solve(3*x**2 - 2*x - 5, x))
```

```python
x = np.arange(-3, 3, 0.1)

plt.axhline(y=0, color='k', alpha=.1)
plt.axvline(x=0, color='k', alpha=.2)

plt.title("3*x**2 - 2*x - 5")
plt.plot(x, 3*x**2 - 2*x -5)

plt.annotate("Solución (-1, 0)", (-3, 1.5), alpha=.5)
plt.annotate("Solución (5/3, 0)", (0, 1.5), alpha=.5)

if len(solc) > 0:
    plt.scatter(solc, np.array(solc)*0, color='black')
```

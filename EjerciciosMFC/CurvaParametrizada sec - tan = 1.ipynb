```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
""" Curva de nivel """

"""Dominio de la Curva"""
x = np.linspace(-9, 9, 50)
y = np.linspace(-9, 9, 50)

x, y = np.meshgrid(x, y)

"""Se dibuja el contorno"""
fig, ax = plt.subplots()

cnt = ax.contour(x, y, y**2 - x**2, [1])
ax.clabel(cnt, inline=True, fontsize=10)

"""Se dibujan los ejes"""
ax.axhline(0, alpha=.2)
ax.axvline(0, alpha=.3)

ax.set_title("Curva de nivel y**2 - x**2")
plt.show()
```

```python
"""Curva parametrizada"""

def x(t):
    return np.tan(t)

def y(t):
    return 1/np.cos(t)

"""Intervalo, se divide en dos para evitar la divisi√≥n por 0"""
t1 = np.linspace(-np.pi/2+.1, np.pi/2-.1, 50)
plt.plot(x(t1), y(t1), c="b")

t2 = np.linspace(np.pi/2+.1, 3*np.pi/2-.1, 50)
plt.plot(x(t2), y(t2), c="b")
plt.title("Curva parametrizada: (tan(t), sec(t))")

"""Ejes"""
plt.axhline(0, alpha=.2)
plt.axvline(0, alpha=.3)
```

```python
""" Otra forma de definir las funciones """
f = lambda t: np.tan(t)
g = lambda t: 1/np.cos(t)

"""Intervalo, se divide en dos para evitar la divisi√≥n por 0"""
t1 = np.linspace(-np.pi/2+.1, np.pi/2-.1, 50)
plt.plot(f(t1), g(t1), c="b")

t2 = np.linspace(np.pi/2+.1, 3*np.pi/2-.1, 50)
plt.plot(f(t2), g(t2), c="b")
plt.title("Curva parametrizada: (tan(t), sec(t))")

"""Ejes"""
plt.axhline(0, alpha=.2)
plt.axvline(0, alpha=.3)
```

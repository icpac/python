```python
import sympy as sym
import math
```

```python
def f(x):
    return x**4 + x**3

x = sym.symbols('x')

def fprime(x):
    return sym.diff(f(x),x)

print("funci贸n: ") 
print(f(x)) 
print("\nderivada: ") 
print(fprime(x)) 

DerivativeOfF = sym.lambdify((x),fprime(x), "numpy")

print("\nValor de la derivada en x = 1:")
print(DerivativeOfF(1))
```

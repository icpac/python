```python
"""
f = e^t    R a R
g = -t^3 + 3t^2 + t  de R a R
g(f) de R a R
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""g sobreyectiva """ 

t = np.arange(-10, 10, 0.01)

def funcion(t):
    return -t**3+3*t**2+t

plt.axvline(x=0, color='k', alpha=.15)
plt.axhline(y=0, color='k', alpha=.15)
plt.plot(t, funcion(t), color="g")
```

```python
""" f inyectiva """
t = np.arange(-30, 0, 0.01)

def funcion(t):
    return np.e**t

plt.axvline(x=0, color='k', alpha=.15)
plt.axhline(y=0, color='k', alpha=.15)
plt.plot(t, funcion(t), color="g")
```

```python
"""g(f) no es inyectiva ni sobreyectiva"""
t = np.arange(-2, 2.5, 0.01)
s = np.arange(-2, 3, 0.01)
s1 = np.arange(-30, 1.5, 0.1)
s2 = np.arange(.5, 4, 0.1)


def funcion(t):
    return np.e**t

def funG(t):
    return -t**3+3*t**2+t

plt.axvline(x=0, color='k', alpha=.15)
plt.axhline(y=0, color='k', alpha=.15)
plt.plot(t, funcion(t), label="f", color="g")
plt.plot(s, funG(s), label="f", color="g")
plt.plot(s1, funG(funcion(s1)), label="f", color="r")
```

```python
"""f de R a R que no es sobreyectiva"""

t = np.arange(-27, 27, .01)
def funcion(t):
    return np.arctan(t)*4/3

plt.axvline(x=0, color='k', alpha=.15)
plt.axhline(y=0, color='k', alpha=.15)
plt.plot(t, funcion(t), label="f", color="g")
```

```python
"""g de R a R que no es inyectiva"""

s = np.arange(-1.97, 1.98, .01)
s1 = np.arange(2.03, 10, .01) 
s2 = np.arange(-10, -2.03, .01) 


def funG(t):
    return t**3 / (t**2-4)

def fG(t):
    return funcion(funG(t))

def Gf(t):
    return funG(funcion(t))

plt.axvline(x=0, color='k', alpha=.15)
plt.axhline(y=0, color='k', alpha=.15)
plt.plot(s, funG(s), label="g", color="g")
plt.plot(s1, funG(s1), label="g", color="g")
plt.plot(s2, funG(s2), label="g", color="g")
plt.scatter(2, 0)
plt.scatter(-2, 0)
```

```python
""" g(f) que es biyectiva"""

t = np.arange(-5, 5, .001)

plt.axvline(x=0, color='k', alpha=.1)
plt.axhline(y=0, color='k', alpha=.1)
plt.plot(t, Gf(t), label="gf", color="r")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#intervalo
x = np.arange(-np.pi, np.pi, .01)

#aproximaciones
coseno1 = lambda t: 1 + 0*t 
coseno2 = lambda t: 1 - (t**2 / 2)  
coseno3 = lambda t: 1 - (t**2 / 2) + (t**4/math.factorial(4))
coseno4 = lambda t: 1 - (t**2 / 2) + (t**4/math.factorial(4)) - (t**6 / math.factorial(6))
coseno5 = lambda t: 1 - (t**2 / 2) + (t**4/math.factorial(4)) - (t**6 / math.factorial(6)) \
+ (t**8/ math.factorial(8))

#Se dibujan las aproximaciones 
plt.plot(x, coseno1(x), label="y = coseno", color="blue", alpha=.1)
plt.plot(x, coseno2(x), label="y = coseno", color="blue", alpha=.2)
plt.plot(x, coseno3(x), label="y = coseno", color="blue", alpha=.3)
plt.plot(x, coseno4(x), label="y = coseno", color="blue", alpha=.4)
plt.plot(x, coseno5(x), label="y = coseno", color="blue", alpha=.5)

#Se dibuja la función coseno
plt.plot(x, np.cos(x), color="r", alpha=.6)

#Etiquetas de las funciones
plt.text(-3, 1, 'y = 1', alpha=.5)
plt.text(-3, -4, 'y = 1-x**2/2', alpha=.5)
plt.text(-3, 0, 'y = 1-x**2/2 + x**4/4!', alpha=.5)
plt.text(2.5, -1.5, 'y = 1-x**2/2 + x**4/4!-x**6/6!', alpha=.5)


plt.title("coseno vs taylor")
plt.show()
```

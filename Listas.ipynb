```python
letters = ['d', 'a', 'e', 'c', 'b']
print ("Lista: ", letters)
```

```python
# sort modifica el orden en la lista
letters.sort()
print ("Ordenada: ", letters)
```

```python
# orden descendente
letters.sort (reverse = True)
print("Invertida: ", letters)
```

```python
original_list = ['Tom', 'James', 'Sarah', 'Fred']

#se hace una copia de la lista
new_list = original_list[:]
new_list.sort()

print("Original: \t", original_list)
print("Nueva ordenada: ", new_list)
```

```python
originallist = ['Tom', 'James', 'Sarah', 'Fred']

#apunta a la lista original
newlist = originallist
newlist.sort()

print("Original:  \t", originallist)
print("'Nueva' ordenada: ", newlist)
```

```python
original = [5, 2, 3, 1, 4]
newer = sorted(original)

print("Original: \t\t", original)
print("Nueva ordenada: ", newer)
```
